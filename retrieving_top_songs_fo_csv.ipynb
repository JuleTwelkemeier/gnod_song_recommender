# Retrieving music information from "DECADE-END CHARTS HOT 100 SONGS"
https://www.billboard.com/charts/decade-end/hot-100
    

```python
#libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
import difflib #for typo control
```

### Retrieving music information from "DECADE-END CHARTS HOT 100 SONGS"<a class="anchor" id="3"></a>

```python
url_decade= "https://www.billboard.com/charts/decade-end/hot-100"
```

```python
#download html with a get request
response = requests.get(url_decade)
response.status_code # 200 status code means OK! when it starts with 4 or 5 it is an error
```

```python
# parse html (create the 'soup')
soup_decade = BeautifulSoup(response.content, "html.parser")
#check that the html code looks like it should
#soup_decade
```

```python

# song_title
song_title_decade = []
for i in soup_decade.find_all("div", {"class":"ye-chart-item__title"}):
    song_title_decade.append(i.text.strip())
#song_title_decade
```

```python
#artist
song_artist_decade = []
for i in soup_decade.find_all("div", {"class":"ye-chart-item__artist"}):
    song_artist_decade.append(i.text.strip())

#song_artist_decade 
```

```python
#peak time of the songs
song_peak_decade = []
for i in soup_decade.find_all("span", {"class":"decade-end-chart-item__peak-info-date"}):
    song_peak_decade.append(i.text.strip("Peak Date"))

#song_peak_decade

```

```python
#creating df from lists
Top100Decade_scraped = pd.DataFrame(list(zip(song_title_decade, song_artist_decade, song_peak_decade)),columns =['Title','Artist', "Peak"],index=pd.RangeIndex(start=1, stop=101))
                     
               
```

```python

Top100Decade_scraped
```

```python
# save the decade-end 2010s to a csv file, as it won't change and therfore we do not need to scrape it everytime
Top100Decade_scraped.to_csv("Top100Decaden.csv", index = False)
```

## Retrieving "1000 Greatest Rock Songs of All Time"
from https://www.rocknrollamerica.net/Top1000.html

```python
url_1000= "https://www.rocknrollamerica.net/Top1000.html"
```

```python
#download html with a get request
response = requests.get(url_1000)
response.status_code # 200 status code means OK! when it starts with 4 or 5 it is an error
```

```python
# parse html (create the 'soup')
soup_1000 = BeautifulSoup(response.content, "html.parser")
#check that the html code looks like it should
#soup_1000
```

```python
#getting the songs using the tag name td and Artist
songs_rock = []

start_row = soup_1000.find(lambda tag: tag and tag.name == "td" and "Artist" in tag.get_text(strip=True)).find_parent("tr")

for row in start_row.find_next_siblings("tr"):
    cells = row.find_all("td")
    
    songs_rock.append(cells[1].get_text())
```

```python
#cleaning the list
songs_rock = [s.replace("\r\n\t\t\t", "") for s in songs_rock]
songs_rock = [s.replace("\xa0", "") for s in songs_rock]

songs_rock
```

```python
#getting the Artists using the tag name td and Artist
artists_rock = []

start_row = soup_1000.find(lambda tag: tag and tag.name == "td" and "Artist" in tag.get_text(strip=True)).find_parent("tr")

for row in start_row.find_next_siblings("tr"):
    cells = row.find_all("td")
    
    artists_rock.append(cells[2].get_text())


```

```python
#cleaning the list
artists_rock = [s.replace("\r\n\t\t\t", "") for s in artists_rock]
artists_rock = [s.replace("\r\n  \t\t\t", "") for s in artists_rock]

artists_rock
```

```python
#creating df from lists
Top1000Rock_scraped = pd.DataFrame(list(zip(songs_rock, artists_rock)),columns =['Title','Artist'],index=pd.RangeIndex(start=1, stop=1001))
                     
```

```python
Top1000Rock_scraped
```

```python
# save the top 1000 rock songs to a csv file, as it won't change and therfore we do not need to scrape it everytime
Top1000Rock_scraped.to_csv("Top1000Rock.csv", index = False)
```
